```python
import pandas as pd
import numpy as np
original_df = pd.read_csv('gwas_4_alz_original_results_final.csv')
results_df = pd.read_csv('gwas_4_alz_intermediate_files/merged_result_df.csv')
results_df = pd.read_csv('gwas_4_alz_result_files/filtered_snps_sd=0.0.csv')


original_df.head()
```

```python
original_df[['chr', 'pos']] = original_df['Position (GRCh37)'].str.split(':', expand=True)
original_df['pos'] = pd.to_numeric(original_df['pos'])
original_df['left_border'] = np.maximum(original_df['pos'] - 500000,0)
original_df['right_border'] = np.maximum(original_df['pos'] + 500000,0)
original_df
```

```python
def count_overlaps(df1, df2):
        
        overlap_count = 0
    
        # Iterate over all pairs of intervals
        for _, row1 in df1.iterrows():
            chr1 = int(row1['chr'])
            for _, row2 in df2.iterrows():
                chr2 = int(row2['chr'])
                # Check if the intervals overlap on the same chromosome
                if chr1==chr2 and row1['right_border'] >= row2['left_border'] and row1['left_border'] <= row2['right_border']:
                    overlap_count += 1
                    break
        return overlap_count

overlap_count = count_overlaps(results_df, original_df)
overlap_count
```
